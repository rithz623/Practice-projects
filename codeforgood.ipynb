```python
# AI Sleep Scheduler â€“ final

import datetime
import random

def input_daily_schedule():
    """Collects user data on sleep, stress, and tasks."""
    print("\n--- Daily Schedule Input ---")
    sleep_hours = float(input("Average hours of sleep last night: "))
    tasks = int(input("Number of major tasks today: "))
    stress = int(input("Stress level (1-10): "))
    bedtime = input("Planned bedtime (HH:MM, 24hr format): ")
    return sleep_hours, tasks, stress, bedtime

def analyze_sleep_balance(sleep_hours, tasks, stress):
    """Computes a balance score combining rest, workload, and stress."""
    balance_score = sleep_hours * 10 - (tasks * 2 + stress * 3)
    return balance_score

def calculate_sleep_efficiency(sleep_hours, stress, tasks):
    """Estimates sleep efficiency percentage and level."""
    efficiency = (sleep_hours / 8) * 100
    efficiency -= stress * 2
    efficiency -= (tasks - 3) * 3 if tasks > 3 else 0
    if efficiency > 80:
        level = "Excellent"
    elif efficiency > 60:
        level = "Moderate"
    else:
        level = "Low"
    return efficiency, level

def recommend_schedule(balance_score, bedtime):
    """Suggests adjusted bedtime or confirms healthy routine."""
    print("\n--- Smart Schedule Recommendation ---")
    if balance_score >= 50:
        print("âœ… Youâ€™re maintaining a healthy schedule! Keep your routine consistent.")
    elif balance_score >= 30:
        print("ðŸ˜´ You may be overworking. Try moving bedtime 30 minutes earlier.")
    else:
        print("âš  High fatigue detected! Re-schedule tasks and rest earlier.")
        try:
            bt = datetime.datetime.strptime(bedtime, "%H:%M")
            new_bt = (bt - datetime.timedelta(minutes=45)).time()
            print(f"ðŸ• Suggested new bedtime: {new_bt.strftime('%H:%M')}")
        except:
            print("Unable to adjust bedtime automatically (format error).")

def wellness_tip(score):
    """Prints a random wellness tip depending on the balance score."""
    tips_good = [
        "Keep your bedtime consistent!",
        "Stay hydrated â€” it improves sleep quality.",
        "Light stretching before bed can help you fall asleep faster."
    ]
    tips_bad = [
        "Take a 20-minute power nap today.",
        "Try 5 minutes of deep breathing before bed.",
        "Avoid screens for an hour before sleeping."
    ]
    print("\nðŸ’¡ Wellness Tip:")
    print(random.choice(tips_good if score >= 40 else tips_bad))

def main():
    print("=== AI Sleep Scheduler ===")
    print("Aligned with SDG 3 â€“ Good Health & Well-Being")
    while True:
        print("\nMenu:\n1. Enter schedule data\n2. Exit")
        choice = input("Choose an option: ")
        if choice == "1":
            sh, t, s, bt = input_daily_schedule()
            score = analyze_sleep_balance(sh, t, s)
            eff, level = calculate_sleep_efficiency(sh, s, t)

            print(f"\nSleep Efficiency: {eff:.1f}% ({level})")
            recommend_schedule(score, bt)
            wellness_tip(score)

        elif choice == "2":
            print("Goodbye! Remember: consistency builds healthy sleep.")
            break
        else:
            print("Invalid choice, please select 1 or 2.")

if __name__ == "__main__":
    main()
```
