```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
trainData = pd.read_csv('/content/drive/MyDrive/MLClass2024/train.csv')
testData = pd.read_csv('/content/drive/MyDrive/MLClass2024/test.csv')
ansData = pd.read_csv('/content/drive/MyDrive/MLClass2024/gender_submission.csv')
trainData.head()
```

```python
testData.head()
```

```python
trainData.drop("Name", axis=1, inplace=True)
testData.drop("Name", axis=1, inplace=True)
trainData.drop("Cabin", axis=1, inplace=True)
testData.drop("Cabin", axis=1, inplace=True)
trainData.drop("Ticket", axis=1, inplace=True)
testData.drop("Ticket", axis=1, inplace=True)
```

```python
trainData
```

```python
encoder = {"Sex": {"male": 0, "female": 1}, "Embarked": {"S": 0, "C": 1, "Q": 2}}
trainData.replace(encoder, inplace=True)
testData.replace(encoder, inplace=True)
```

```python
trainData
```

```python
testData
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X = trainData.drop("Survived", axis=1)
y = trainData["Survived"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
ansData
```

```python
trainedData = RandomForestClassifier()
trainedData.fit(X_train, y_train)
```

```python
yPredictions = trainedData.predict(X_test)
yPredictions[:10]
```

```python
trainedData.score(X_test, y_test)
```

```python
randomSearchParams = {"n_estimators": [10, 50, 100, 500, 1000],
                      "n_jobs": [2, 5, 10, 50, 100],
                      "min_samples_leaf": [1, 2, 5, 10, 50],
                      "max_features": [1, 2, 5, 10, 50],
                      "max_depth": [1, 2, 5, 10, 50]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
randomSearchModel = RandomForestClassifier(n_jobs=-1)
rs_randomSearchModel = RandomizedSearchCV(estimator = RandomForestClassifier(n_jobs=-1),
                                          param_distributions = randomSearchParams,
                                          n_iter = 30,
                                          cv = 5,
                                          verbose = 2)
```

```python
rs_randomSearchModel.fit(X_train, y_train)
```

```python
rs_randomSearchModel.best_params_
```

```python
randomYPredictions = rs_randomSearchModel.predict(X_test)
randomYPredictions[:10]
```

```python
randomSearchScore = rs_randomSearchModel.score(X_test, y_test)
randomSearchScore
```
