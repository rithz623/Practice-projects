```python
heart_disease = pd.read_csv("/content/content.drive/MyDrive/MLClass2024/heart-disease (4).csv")
heart_disease.head(),"", "Missing Values", heart_disease.isna().sum(), "",
```

```python
X = heart_disease.drop("target", axis=1)
y= heart_disease.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=1500, n_jobs=500, max_depth=1009,max_leaf_nodes=98, max_features= 49, max_samples= 100, min_samples_leaf= 23, min_samples_split= 23)

model.fit(X_train, y_train)
```

```python
model.get_params()
model.score(X_test, y_test)
```
